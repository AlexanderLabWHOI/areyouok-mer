```python
metadata = pd.read_csv("../data/metadata_kmerproj.csv")
normalstp = metadata[(metadata.Species == "TP") & ((metadata.Treatment == "ref") | (metadata.Treatment == "rep"))]
normalstp
```

```python
colsmat = list(matrix_cmp_21_orig.columns)
targets = [curr for curr in range(len(colsmat)) if (colsmat[curr] in list(normalstp.Name))]
matrix_cmp_21 = matrix_cmp_21_orig.iloc[targets,targets]

colsmat = list(matrix_cmp_31_orig.columns)
targets = [curr for curr in range(len(colsmat)) if (colsmat[curr] in list(normalstp.Name))]
matrix_cmp_31 = matrix_cmp_31_orig.iloc[targets,targets]

colsmat = list(matrix_cmp_51_orig.columns)
targets = [curr for curr in range(len(colsmat)) if (colsmat[curr] in list(normalstp.Name))]
matrix_cmp_51 = matrix_cmp_51_orig.iloc[targets,targets]
```

```python
fig, axs = plt.subplots(1, 3, sharey= True, figsize=(35,7))
                        #figsize=(80,20)) this is for giant figure
ax1 = axs[0]
im = ax1.imshow(matrix_cmp_21, interpolation='nearest',aspect = 'auto')
im.set_clim(0,1)
ax1.set_title('k = 21',size=20)
for i in range(len(matrix_cmp_21.index)):
    for j in range(len(matrix_cmp_21.columns)):
        text = ax1.text(j, i, round(matrix_cmp_21.iloc[i, j],4),
                       ha="center", va="center", color="w")
        
        
ax2 = axs[1]
ax3 = axs[2]
ax2.set_title('k = 31',size=20)        
im = ax2.imshow(matrix_cmp_31, interpolation='nearest', aspect = 'auto')
im.set_clim(0,1)
for i in range(len(matrix_cmp_31.index)):
    for j in range(len(matrix_cmp_31.columns)):
        text = ax2.text(j, i, round(matrix_cmp_31.iloc[i, j],4),
                       ha="center", va="center", color="w")
labels = list(matrix_cmp_21.columns)
ax1.set_xticks(np.arange(len(labels)))
ax1.set_yticks(np.arange(len(labels)))
ax2.set_xticks(np.arange(len(labels)))
ax3.set_xticks(np.arange(len(labels)))

ax1.set_yticklabels(labels)
ax1.set_xticklabels(labels)
ax2.set_xticklabels(labels)
ax3.set_xticklabels(list(matrix_cmp_21.columns))

# Rotate the tick labels and set their alignment.
plt.setp(ax2.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
plt.setp(ax1.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
plt.setp(ax3.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
ax1.tick_params(labelsize=15)
ax2.tick_params(labelsize=15)
ax3.tick_params(labelsize=15)
        
ax3.set_title('k = 51',size=20)        
im = ax3.imshow(matrix_cmp_51, interpolation='nearest',aspect = 'auto')
im.set_clim(0,1)
for i in range(len(matrix_cmp_51.index)):
    for j in range(len(matrix_cmp_51.columns)):
        text = ax3.text(j, i, round(matrix_cmp_51.iloc[i, j],3),
                       ha="center", va="center", color="w")
        
fig.colorbar(im, ax=axs, shrink=0.6)        
#ax3.colorbar()
plt.savefig(fname = "nutrientrepleteplusreference.pdf",transparent=True)
```

```python
def makefigure(matrix_cmp_21,matrix_cmp_31,matrix_cmp_51,figname,dims,ftsz=8,titlesize=20):
    fig, axs = plt.subplots(1, 3, sharey= True, figsize=dims)
                        #figsize=(35,7)) this is for small figures
                        #figsize=(80,20)) this is for giant figure
    ax1 = axs[0]
    im = ax1.imshow(matrix_cmp_21, interpolation='nearest',aspect = 'auto')
    im.set_clim(0,1)
    ax1.set_title('k = 21',size=titlesize)
    for i in range(len(matrix_cmp_21.index)):
        for j in range(len(matrix_cmp_21.columns)):
            text = ax1.text(j, i, round(matrix_cmp_21.iloc[i, j],4),
                           ha="center", va="center", color="w",fontsize=ftsz)


    ax2 = axs[1]
    ax3 = axs[2]
    ax2.set_title('k = 31',size=titlesize)        
    im = ax2.imshow(matrix_cmp_31, interpolation='nearest', aspect = 'auto')
    im.set_clim(0,1)
    for i in range(len(matrix_cmp_31.index)):
        for j in range(len(matrix_cmp_31.columns)):
            text = ax2.text(j, i, round(matrix_cmp_31.iloc[i, j],4),
                           ha="center", va="center", color="w",fontsize=ftsz)
    labels = list(matrix_cmp_21.columns)
    ax1.set_xticks(np.arange(len(labels)))
    ax1.set_yticks(np.arange(len(labels)))
    ax2.set_xticks(np.arange(len(labels)))
    ax3.set_xticks(np.arange(len(labels)))

    ax1.set_yticklabels(labels)
    ax1.set_xticklabels(labels)
    ax2.set_xticklabels(labels)
    ax3.set_xticklabels(list(matrix_cmp_21.columns))

    # Rotate the tick labels and set their alignment.
    plt.setp(ax2.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")
    plt.setp(ax1.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")
    plt.setp(ax3.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")
    ax1.tick_params(labelsize=15)
    ax2.tick_params(labelsize=15)
    ax3.tick_params(labelsize=15)

    ax3.set_title('k = 51',size=titlesize)        
    im = ax3.imshow(matrix_cmp_51, interpolation='nearest',aspect = 'auto')
    im.set_clim(0,1)
    for i in range(len(matrix_cmp_51.index)):
        for j in range(len(matrix_cmp_51.columns)):
            text = ax3.text(j, i, round(matrix_cmp_51.iloc[i, j],3),
                           ha="center", va="center", color="w",fontsize=ftsz)

    fig.colorbar(im, ax=axs, shrink=0.6)        
    #ax3.colorbar()
    plt.savefig(fname = figname,transparent=True)
    
def selectspecific(targetnames):
    colsmat = list(matrix_cmp_21_orig.columns)
    targets = [curr for curr in range(len(colsmat)) if (colsmat[curr] in list(targetnames.Name))]
    matrix_cmp_21 = matrix_cmp_21_orig.iloc[targets,targets]

    colsmat = list(matrix_cmp_31_orig.columns)
    targets = [curr for curr in range(len(colsmat)) if (colsmat[curr] in list(targetnames.Name))]
    matrix_cmp_31 = matrix_cmp_31_orig.iloc[targets,targets]

    colsmat = list(matrix_cmp_51_orig.columns)
    targets = [curr for curr in range(len(colsmat)) if (colsmat[curr] in list(targetnames.Name))]
    matrix_cmp_51 = matrix_cmp_51_orig.iloc[targets,targets]
    return matrix_cmp_21,matrix_cmp_31,matrix_cmp_51
```

```python
metadata = pd.read_csv("../data/metadata_kmerproj.csv")
targetnames = metadata[(metadata.Species == "TP") & (metadata.Treatment == "sil")]
targetnames
```

```python
matrix_cmp_21,matrix_cmp_31,matrix_cmp_51 = selectspecific(targetnames)
```

```python
makefigure(matrix_cmp_21,matrix_cmp_31,matrix_cmp_51,"SilLimited.pdf",(35,7))
```
